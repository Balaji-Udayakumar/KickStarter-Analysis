## Loading the amount pledged (in U.S. dollars) data across all projects

- We are loading data from'HW3_kickstarter.csv'.
- The `np.loadtxt` function is used to load the data from the file.
- `skiprows=1` skips the header row when reading the data.
- `usecols=11` specifies that we are fetching the data from the 11th column (the index starts from 0).
- `delimiter=','` indicates that the data in the CSV file is separated by commas.

```python
# Loading the amount pledged (in U.S. dollars) data
amount_pledged = np.loadtxt('HW3_kickstarter.csv', skiprows=1, usecols = 11, delimiter=',')
```

## Displaying descriptive statistics

- We calculate various statistics related to the amount pledged (in U.S. dollars) for Kickstarter projects.
- The total number of projects is calculated using `len(amount_pledged)`.
- Minimum, maximum, mean, median, and standard deviation for the amount pledged are calculated using NumPy functions.
- The percentage of projects that earned total pledges of at least $2,500 is calculated and printed.
- Finally, the results are being printed.

```python
# Calculating total number of projects

total_no_of_projects = len(amount_pledged)
total_no_of_projects

# Calculating min, max, mean, median, and standard deviation for amount pledged in US dollars

min_amt_pledged = np.min(amount_pledged)
max_amt_pledged = np.max(amount_pledged)
mean_amt_pledged = np.mean(amount_pledged)
median_amt_pledged = np.median(amount_pledged)
std_dev_amt_pledged = np.std(amount_pledged)

# Calculating % of projects that earned total pledges of at least $2500.

pledge_limit = 2500
projects_above_limit = np.sum(amount_pledged >= pledge_limit)
percentage_above_limit = (projects_above_limit/total_no_of_projects) * 100

# Print results

print("--------Total number of projects----------")
print("\t\t",total_no_of_projects)
print("--------Amount pledged (in U.S. dollars): min, max, mean, median, and standard deviation----------")
print("\t\tMinimum:", min_amt_pledged)
print("\t\tMaximum:", "{:.2f}".format(max_amt_pledged))
print("\t\tMean:", "{:.2f}".format(mean_amt_pledged))
print("\t\tMedian:", "{:.2f}".format(median_amt_pledged))
print("\t\tStandard Deviation:", "{:.2f}".format(std_dev_amt_pledged))
print("--------Percentage of projects that earned total pledges of at least $2,500.----------")
print("\t\t", "{:.2f}".format(percentage_above_limit))
```

## Load the project categories data

- We are loading data from 'HW3_kickstarter.csv'.
- The `np.loadtxt` function is used to load the data from the file.
- `skiprows=1` skips the header row when reading the data.
- `usecols=12` specifies that we are fetching the data from the 12th column (the index starts from 0).
- `delimiter=','` indicates that the data in the CSV file is separated by commas.

```python
## Loading the project categories data
project_category = np.loadtxt('HW3_kickstarter.csv', skiprows=1, usecols = 12, delimiter=',', dtype = str)
project_category
```

## Calculating frequency of each category

- An empty dictionary called `frequency_of_category` is created to determine the frequency of each category.
- A loop iterates through the `project_category` array iterating through the categories of projects.
- Inside the loop, we check if a category is already in the `frequency_of_category` dictionary. If it is, we increment the count for that category; otherwise, we initialize the count to 1.
- Finally, we print the frequency of each category along with the category name using a loop.

```python
# Created an empty dict to determine the frequency of each category
frequency_of_category = {}

# Code Logic to count the frequency of each category
for category in project_category:
    if category in frequency_of_category:
        frequency_of_category[category]+=1
    else:
        frequency_of_category[category]=1

# Printing frequency of each category
for category,count in frequency_of_category.items():
    print(f'{category}: {count}')
```

## Calculating the proportion (percentage) of observations for each category

- A dictionary comprehension is used to calculate the proportion (percentage) of each category in the dataset.
- `proportion_of_category` is created as a dictionary where each category is the key, and the corresponding proportion (percentage) is calculated using the formula `(count / total_no_of_projects) * 100`.
- A loop is used to print the category and its corresponding percentage from the `proportion_of_category` dictionary.

```python
# Using a dictionary comprehension to calculate the proportion(percentage) of categories
proportion_of_category = {category:(count/total_no_of_projects)*100 for category,count in frequency_of_category.items()}

# Printing the proportion(percentage) of categories
for category,category_percentage in proportion_of_category.items():
    print(f'{category}: {category_percentage}')
```

## Determining the total number of projects identified as staff picks

### Finding the frequency of each of staff pick's outcomes {TRUE,FALSE}

- An empty dictionary named `project_staff_pick_count` is created to count the frequency of different values in the `project_staff_pick` array.
- Next we iterate through the `project_staff_pick` array and for each `staff_pick` value, we check if that value already exists in the `project_staff_pick_count` dictionary. If it does, we retrieve the current count; otherwise, we defaults it to 0.
- The code then increments the count for the `staff_pick` value in the dictionary.
- Finally, the resulting dictionary `project_staff_pick_count` is displayed, which shows the count of each unique `staff_pick` outcome (True,False) and how many times it appears in the list.

```python
# Creating an empty dictionary to count the frequency of staff pick outcomes (True,False)
project_staff_pick_count = {}
# Iterating through the project_staff_pick array
for staff_pick in project_staff_pick:
    # Checking if the staff_pick outcome already exists in the dictionary
    # If it does, get the current count; otherwise, default to 0
    count = project_staff_pick_count.get(staff_pick,0)
    # Incrementing the count for the staff_pick value
    project_staff_pick_count[staff_pick] = count + 1
# Displaying the resulting staff pick count dictionary
project_staff_pick_count
```

## Similar steps are followed for finding the frequency of each of spotlight's outcomes {TRUE,FALSE}

```python
# Creating an empty dictionary to count the frequency of spotlight outcomes (True,False)
project_spotlight_count = {}
# Iterating through the project_spotlight array
for spotlight in project_spotlight:
    # Checking if the spotlight outcome already exists in the dictionary
    # If it does, get the current count; otherwise, default to 0
    count = project_spotlight_count.get(spotlight,0)
    # Incrementing the count for the spotlight value
    project_spotlight_count[spotlight] = count + 1
# Displaying the resulting spotlight count dictionary
project_spotlight_count
```

## Determining the total number of projects identified as staff picks ( i.e. when staff_pick is 'TRUE')

```python
# Iterate through the project_staff_pick_count dictionary to find the count of projects featured as staff picks
for category,count in project_staff_pick_count.items():
    # Check if the category is 'TRUE', indicating that the project is featured as a staff pick
    if category == 'TRUE':
        # Assign the count of staff-picked projects to the variable 'total_staff_picks'
        total_staff_picks = count
        # Print the total number of projects featured as staff picks
        print(f'The total number of projects featured as staff picks is: {count}')
```

## Determining the total number of projects identified as spotlights ( i.e. when spotlight is 'TRUE')

```python
# Iterate through the project_spotlight_count dictionary to find the count of projects featured as spotlights
for category,count in project_spotlight_count.items():
    # Check if the category is 'TRUE', indicating that the project is featured as a spotlight
    if category == 'TRUE':
        # Assign the count of spotlight projects to the variable 'total_spotlights'
        total_spotlights = count
        # Print the total number of projects featured as spotlights
        print(f'The total number of projects featured as spotlights is: {count}')
```

## Finding the success rates for staff-picked and spotlight projects

- Two counters, `staff_pick_success_count` and `spotlight_success_count`, are initialized to keep track of the number of staff-picked and spotlighted successful projects, respectively.
- A loop iterates through the range of projects (`total_no_of_projects`) to process each project.
- Within the loop, it checks if the project is 'successful' based on the project state.
- If a project is successful, it checks whether it was featured as a staff pick (`project_staff_pick`) and whether it was spotlighted (`project_spotlight`).
- If a project meets the criteria for being staff-picked or spotlighted, the respective counters are incremented.
- Success rates for staff-picked and spotlighted projects are calculated as percentages.
- Finally, the success rates are printed to the console.

```python
# Initialize counters for staff-picked and spotlighted successful projects
staff_pick_success_count = 0
spotlight_success_count = 0

# Iterate through the range of projects
for itr in range(total_no_of_projects):
    # Check if the project is 'successful'
    if project_states[itr] == 'successful':
        # Check if the project was featured as a staff pick
        if project_staff_pick[itr] == 'TRUE':
            # Increment the count of staff-picked successful projects
            staff_pick_success_count+=1
        # Check if the project was spotlighted
        if project_spotlight[itr] == 'TRUE':
            # Increment the count of spotlighted successful projects
            spotlight_success_count+=1
            
# Calculating the success rates for staff-picked and spotlighted projects
staff_pick_success_rate = (staff_pick_success_count/total_staff_picks)*100
spotlight_success_rate = (spotlight_success_count/total_spotlights)*100

# Printing the success rates
print(f'Success rate of staff picked projects is {staff_pick_success_rate:.2f}')
print(f'Success rate of spotlight projects is {spotlight_success_rate:.2f}')
```

## 1. Finding Distribution of Projects by country

- A dictionary named `project_by_country` is created to store counts of projects by each country. It iterates through the `country` data.
- A threshold value is defined as `threshold`.
- A variable `others_sum` is initialized to hold the sum of values below the threshold.
- A new dictionary `final_project_by_country` is created to accommodate counts above or equal to the threshold, and the `Others` category.
- The code iterates through the original `project_by_country` dictionary and adds values below the threshold to `others_sum` while adding values above or equal to the threshold to `final_project_by_country`.
- Finally, the `Others` category with the sum of values below the threshold is added to `final_project_by_country` dictionary, and the result is printed.

```python
# creating project_by_country dict to store counts of projects by each country
project_by_country = {}
for each_country in country:
    count = project_by_country.get(each_country,0)
    project_by_country[each_country] = count + 1
# Defining a threshold value
threshold = 79

# Initializing a variable to hold the sum of values below the threshold
others_sum = 0

# Create a new dictionary with the counts and an "Others" category to accomodate the sum of values below the threshold 79
final_project_by_country = {}

# Iterate through the original dictionary
for key,value in project_by_country.items():
    if value<threshold:
        # Add values below the threshold to the 'others_sum'
        others_sum += value
    else:
        # Add values above or equal to the threshold to the final_project_by_country dict
        final_project_by_country[key] = value

# Add the 'Others' category with the sum of values below the threshold
final_project_by_country['Others'] = others_sum

# Printing the final project by country
print(final_project_by_country)
```

## 2. Distribution of proportion of successful projects by country

## Code explanation to find count of successful projects by country

- An empty dictionary `success_by_country` is initialized to count the number of successful projects by country.
- An iterator `itr` is initialized to keep track of the current project.
- A `for` loop iterates through each country in the `country` list.
- Inside the loop, we check if the project state at the current iterator position is 'successful'.
- If the project is successful, it updates the count for that country in the `success_by_country` dictionary.
- After processing all projects, we print the count of successful projects by country along with the total projects for that country using`project_by_country` dictionary.

```python
# Initializing an empty dictionary to count successful projects by country
success_by_country = {}

# Initializing an iterator to keep track of the current project
itr = 0

# Iterating through each country in the 'country' list
for each_country in country:
    # Checking if the project state at the current iterator position is 'successful'
    if project_states[itr] == 'successful':
        # If the country is already in the 'success_by_country' dictionary, we increment the count
        if each_country in success_by_country:
            success_by_country[each_country] += 1
        # If the country is not in the dictionary, we add it with a count of 1
        else:
            success_by_country[each_country] = 1
    # Incrementing the iterator to move to the next project
    itr = itr + 1

# Printing the count of successful projects by country
for key, value in success_by_country.items():
    # Assuming you have a 'project_by_country' dictionary, you can print the total projects for that country
    print(f'Country - {key}: Success count - {value}: total projects - {project_by_country[key]}')
```

## Calculating the percentage of project successes for each country and storing it in a dictionary

```python
# Using a dictionary comprehension to calculate the proportion of project successes by country
proportion_of_success_projects_by_country = {key:(value/project_by_country[key])*100 for key,value in success_by_country.items()}

proportion_of_success_projects_by_country
```
